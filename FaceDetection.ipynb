```
import cv2
from google.colab.patches import cv2_imshow
```

```
classificador = cv2.CascadeClassifier('/content/haarcascade_frontalface_default.xml')
image = cv2.imread("/content/pessoas4.jpg")
```

```
gray_scale = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_scale)
```

```
faceDetected = classificador.detectMultiScale(gray_scale, scaleFactor = 1.2, minNeighbors = 10)

for (x,y,l,a) in faceDetected:
  cv2.rectangle(image, (x,y), (x+l,y+a) ,(0,0,255),2)
cv2_imshow(image)
```

# DETECÇÃO DE OLHOS

```
#detectando rosto

#classificador face
classificador2 = cv2.CascadeClassifier('/content/haarcascade_frontalface_default.xml')
image2 = cv2.imread("/content/pessoas4.jpg")
gray_scale2 = cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY)

#classificador olhos
classificador_olhos = cv2.CascadeClassifier("/content/haarcascade_eye.xml")
face_detected = classificador2.detectMultiScale(gray_scale2)


for (x,y,l,a) in face_detected:
  cv2.rectangle(image2, (x,y), (x+l, y+a), (100,100,255),2)
  regiao = image2[y:y + a, x: x + l]

  regiaoCinzaOlho = cv2.cvtColor(regiao, cv2.COLOR_BGR2GRAY)
  olhosDetected = classificador_olhos.detectMultiScale(regiaoCinzaOlho)

  #loop dentro da região da face
  for (ox,oy,ol,oa) in olhosDetected:
    cv2.rectangle(regiao, (ox,oy), (ox+ol, oy+oa), (255,0,255),2)

cv2_imshow(image2)
cv2.waitKey()
```
